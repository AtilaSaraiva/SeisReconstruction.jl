```julia
using PyPlot, SeisPlot, SeisReconstruction, SeisProcessing

```

```julia
# Create parabolic events and apply Spitz FX interpolation to interpolate
# the data. By interpolate we mean going from dx sampling to dx/2 sampling.

# Spitz, S., 1991, Seismic trace interpolation in the F-X domain: Geophysics, 56, 785-794.

d = SeisParabEvents(dt=0.004,nt=250,nx1=20,dx1=10,tau=[0.2, 0.4, 0.6],
                                            p1=[5.9,-1.5, -3.9],
                                            amp=[1.0,-1.0,1.0])
dec = SeisSpitzFX(d;dt=0.004,npf=4,pre1=.1,pre2=.5,flow=1,fhigh=90)

figure(1,figsize=(10, 5))
SeisPlotTX(hcat(d,dec),style="wiggles")
title("Spitz FX Interpolation")
gcf()
```
